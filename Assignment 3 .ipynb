```python
from random import choice
import time
```

```python
infinity = 1000000
board = [
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 0],
]

x_choice='x'
o_choice='o'
```

```python
def gameover(board):
        goal=[
        [board[0][0], board[0][1], board[0][2]],
        [board[1][0], board[1][1], board[1][2]],
        [board[2][0], board[2][1], board[2][2]],
        [board[0][0], board[1][0], board[2][0]],
        [board[0][1], board[1][1], board[2][1]],
        [board[0][2], board[1][2], board[2][2]],
        [board[0][0], board[1][1], board[2][2]],
        [board[2][0], board[1][1], board[0][2]],]
        if ['x','x','x'] in goal:
            return 'x'
        elif ['o','o','o'] in goal:
            return 'o'
        else:
            return False 
    
```

```python
def check_space(board):
    #pick up all emptys set
    emptys =[]
    
    for x, row in enumerate(board):
        for y,empty in enumerate(row):
            if empty==0:
                emptys.append([x,y])
    return emptys
def print_board(board):
    
    print("===========")
    for row in board:
        print()
        for cell in row:
            if cell is 'x':
                print('x ',end='')
            elif cell is 'o':
                print('o ',end='')
            else:
                print('* ',end='')
    print()
#print_board(board)
```

```python
def trans(agent):
    if agent==-1:
        return 'o'
    return 'x'
```

```python
def _other(agent):
    if agent == 'x':
        return 'o'
    return 'x'
def minimax_1(board,agent,depth):
    #expect x win
    if agent ==1:
        #row, col ,score
        best=[-1,-1,-infinity]
    else:
        best=[-1,-1,+infinity]
    if depth ==0 or gameover(board) is not False:
        
        if gameover(board) =='x':
            return [-1,-1,1]
            
        elif gameover(board)=='o':
            return [-1,-1,-1]
        else:
            return [-1,-1,0]
    for cell in check_space(board):
        x,y=cell[0],cell[1]
        board[x][y] = trans(agent)
        score= minimax_1(board,-agent,depth-1,)
        board[x][y] =0
        score[0],score[1] =x,y
        
        if agent ==1:
            #max
            if score[2]>best[2]:
                best=score 
        else:
            #min
            if score[2]<best[2]:
                best=score
        
    return best
                    
                    
        
```

```python
def minimax_2(board,agent,depth):
    #expect o will win
    if agent ==-1:
        #row, col ,score
        best=[-1,-1,-infinity]
    else:
        best=[-1,-1,+infinity]
    if depth ==0 or gameover(board) is not False:
        
        if gameover(board) =='x':
            return [-1,-1,-1]
        elif gameover(board)=='o':
            return [-1,-1,1]
        else:
            return [-1,-1,0]
    for cell in check_space(board):
        x,y=cell[0],cell[1]
        board[x][y] = trans(agent)
        score= minimax_2(board,-agent,depth-1,)
        board[x][y] =0
        score[0],score[1] =x,y
        
        if agent ==-1:
            #max
            if score[2]>best[2]:
                best=score 
        else:
            #min
            if score[2]<best[2]:
                best=score
    return best
```

```python
def move(agent):
    #print_board(board)
    if len(check_space(board))==9:
        x=choice([0,1,2])
        y=choice([0,1,2])
    elif agent =='x':
        mi= minimax_1(board,1,len(check_space(board)))
        x,y =mi[0],mi[1]
    elif agent =='o':
        mi= minimax_2(board,-1,len(check_space(board)))
        
        x,y =mi[0],mi[1]
    else: pass
    print("move:",agent,x,y)
    if [x,y] in check_space(board):
        board[x][y]=agent
```

```python
def start():
    #first agent is 1, the opposite is -1
    agent = 'x'
    while True:
        time.sleep(1)
        if gameover(board) is not False:
            break
        if len(check_space(board))<=0:
            break
        move(agent)
        agent=_other(agent)
        print("main agent:",agent)
        print_board(board)
        
        
    if gameover(board) is 'o':
        print("o is winner")
    elif gameover(board) is 'x':
        print("x is winner")
    else:
        print("DRAW")
        
        
```
